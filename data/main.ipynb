```python
import pandas as pd
from math_model.src.graph_generator import GraphGenerator
```

# Reading from csv

```python
df = pd.read_csv('out.csv')
df
```

# Clearing data

```python
df = df[df.Altitude <= 12_500]
df
```

```python
startIndexOffset = 2821
startTimeOffset = 0

for index, item in df.iterrows():
    if index == startIndexOffset + 1:
        startTimeOffset = item[0]
        df.at[index,'Time'] = 0
        continue

    df.at[index,'Time'] = item[0] - startTimeOffset
```

```python
df
```

# Getting values for plots

```python
time_values = df['Time']
alt_values = df['Altitude']
velocity_values = df['Velocity']
sol_fuel_values = df['SolidFuel']
liq_fuel_values = df['LiquidFuel']
angle_values = df['Angle']
time_values
```
